```python
import cv2
import matplotlib.pyplot as plt

def main():
    # Read the image
    _img = cv2.imread('X:\Infosys_STMS\Smart_Traffic_Management_System\Data\Data_Images\img.png', cv2.IMREAD_COLOR)

    # Check if the image is loaded
    if _img is None:
        print("Error: Image not found or unable to load.")
        return

    # Apply blur to the image
    blur_image = cv2.blur(_img, (10, 10))  # (10, 10) is the kernel size

    # Convert images from BGR to RGB for displaying with Matplotlib
    _img_rgb = cv2.cvtColor(_img, cv2.COLOR_BGR2RGB)
    blur_image_rgb = cv2.cvtColor(blur_image, cv2.COLOR_BGR2RGB)

    # Display the original and blurred images using Matplotlib
    plt.figure(figsize=(10, 5))

    # Original image
    plt.subplot(1, 2, 1)
    plt.imshow(_img_rgb)
    plt.title("Original Image")
    plt.axis('off')

    # Blurred image
    plt.subplot(1, 2, 2)
    plt.imshow(blur_image_rgb)
    plt.title("Blurred Image")
    plt.axis('off')

    plt.show()

if __name__ == "__main__":
    main()
```

```python
def main():
    # Read the image in color
    _img = cv2.imread('X:\Infosys_STMS\Smart_Traffic_Management_System\Data\Data_Images\img.png', cv2.IMREAD_COLOR)
    
    if _img is None:
        print("Error: Image not found or path is incorrect.")
        return
    
    # Coordinates for the text position
    text_position = (20, 50)  # (x, y) coordinates for the text

    # Add text to the image
    cv2.putText(
        _img,
        "Vehicles",
        text_position,
        cv2.FONT_HERSHEY_SIMPLEX,
        2.0,  # Font size
        (255, 255, 255),  # Text color (white in BGR)
        2  # Thickness
    )

    # Try displaying with OpenCV
    try:
        cv2.imshow("Traffic Image", _img)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except Exception as e:
        # Use matplotlib as a fallback
        _img_rgb = cv2.cvtColor(_img, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
        plt.imshow(_img_rgb)
        plt.title("Traffic Image with Text")
        plt.axis('off')
        plt.show()

if __name__ == "__main__":
    main()
```

```python
import cv2
import matplotlib.pyplot as plt

def main():
    # Load the image in color (BGR format)
    _img = cv2.imread('X:\Infosys_STMS\Smart_Traffic_Management_System\Data\Data_Images\img.png', cv2.IMREAD_COLOR)

    if _img is None:
        print("Error: Image not found or path is incorrect.")
        return

    # Convert the image to RGB format
    _img_rgb = cv2.cvtColor(_img, cv2.COLOR_BGR2RGB)

    # Try displaying with OpenCV
    try:
        cv2.imshow("Traffic Image in BGR", _img)  # Original BGR image
        cv2.imshow("Traffic Image in RGB", _img_rgb)  # RGB image (for demonstration)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    except Exception as e:
        # Fallback: Display using matplotlib
        plt.imshow(_img_rgb)
        plt.title("Traffic Image in RGB")
        plt.axis('off')  # Hide axis for better visualization
        plt.show()

if __name__ == "__main__":
    main()
```
